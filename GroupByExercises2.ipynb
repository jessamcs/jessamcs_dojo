```python
#Grouping
#Select data in the month of August 2005 having flow id=7
#Group the data based on date and get the max count per date

from datetime import datetime
df['Date'] = pd.to_datetime(df['Date'])
month_filter = df['Date'].dt.month == 8
year_filter = df['Date'].dt.year == 2005
df_months = df.loc[month_filter & year_filter, :]
df_months.sort_values(by=['Count'])

```

```python
df.groupby(['Date']).max()
```
